```python
import pandas as pd
```

```python
#import excel file

df = pd.read_excel(
    'E.coli MG1655 selection glpD protein BLAST.xlsx'
)
df.head()
```

```python
df.iloc[0].Description
```

```python
#write a function to extract the text from the square brackets 

def extract_from_square_brackets(text: str) -> str:
    """
    Returns a substring with the words inside square brackets

    Examples
    --------

    >>> extract_from_square_brackets('hello [world]!')
    'world'

    >>> extract_from_square_brackets('what a marvelous [day] it is.')
    'day'
    
    """
    start = text.index('[') + 1
    end = text.index(']')
    substring = text[start:end]
    return substring
    


extract_from_square_brackets('glycerol-3-phosphate dehydrogenase [Escherichia coli]')
```

```python
# write the text from the square brackets into a new 'Species' column
df['Species'] = df.Description.apply(extract_from_square_brackets)
df.head()
```

```python
df.iloc[0]
```

```python
#write the first name of the Species column into a new 'Genus' column
df['Genus'] = [x.split()[0] for x in df['Species']]
```

```python
df.head()
```

```python
#save the dataframe as an excel column
df.to_excel('GlpD Genus column test.xlsx', index=False)
```

```python
#import the excel file containing the full phylogenetic information for all bacterial genera
dfphylo = pd.read_excel(
    'Full Taxonomie.xlsx'
)
dfphylo.head()
```

```python
#merge the Blast result dataframe and the dataframe containing phylogenetic information for all bacteria by the Genus column and create a new dataframe

df_merged = pd.merge(
    left=df, 
    right=dfphylo[['Genus','Family','Order','Class',]], 
    left_on='Genus', 
    right_on='Genus',
)
df_merged.head()
```

```python
#save the merged dataframe as an excel file
df_merged.to_excel('GlpD final.xlsx', index=False)
```
